```python
import wandb
```

```python
api = wandb.Api(timeout=120)
runs = api.runs("wilrop/MORL-Baselines")
```

```python
for run in runs:
    algo = run.config['algo']
    env_id = run.config['env_id']
    seed = run.config['seed']
    
    for history in run.scan_history(keys=['eval/front', 'global_step']):
        global_step = history['global_step']
        file = run.file(history['eval/front']['path'])
        print('Downloading artifact', file.name)
        file.download(root=f'artifacts/{env_id}/{algo}/{seed}/{global_step}')
    print(f'Finished {env_id} - {algo} - {seed}')
```
